# Error Analysis
Inspect misclassifications and confusion patterns.

```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

df_errors = pd.read_csv('../reports/test_predictions.csv')
misclassified = df_errors[df_errors['true_label'] != df_errors['predicted_label']]
print(f'Total misclassified: {len(misclassified)} / {len(df_errors)} ' \n
      f'({len(misclassified)/len(df_errors)*100:.1f}%)')
```

```python
cm = confusion_matrix(df_errors['true_label'], df_errors['predicted_label'])
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds')
plt.title('Confusion Matrix - Test Set')
plt.savefig('../reports/test_confusion_matrix.png')
```

```python
print('Misclassified sample feature averages:')
print(misclassified[['N','P','K','ph','moisture']].describe())
```

```python
print('Class-wise error rates:')
for crop in df_errors['true_label'].unique():
    subset = df_errors[df_errors['true_label'] == crop]
    err_rate = (subset['true_label'] != subset['predicted_label']).mean()
    print(f'  {crop}: {err_rate*100:.1f}% error rate (n={len(subset)})')
```
