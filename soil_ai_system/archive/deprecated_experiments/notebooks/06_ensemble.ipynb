# Stacked Ensemble
Build a stacked ensemble on top of RF and XGBoost.

```python
from ensemble.stacking import build_stacking_ensemble
from training.evaluate import stratified_split
import pandas as pd
from config import PROCESSED_DATA_PATH, FEATURE_COLS

df = pd.read_csv(f'{PROCESSED_DATA_PATH}merged_soil_data.csv')
available = [c for c in FEATURE_COLS if c in df.columns]
X = df[available].values
y = df['crop'].values
X_train, X_val, X_test, y_train, y_val, y_test = stratified_split(X, y)
build_stacking_ensemble(X_train, y_train, X_test, y_test, task='crop', num_classes=len(set(y)))
```
