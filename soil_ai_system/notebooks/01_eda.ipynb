# EDA - Crop Recommendation Dataset
Review schema, distributions, correlations, and imbalance signals.

```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
df = pd.read_csv('../datasets/raw/crop_recommendation.csv')
print(df.shape, '\n')
print(df.dtypes, '\n')
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12,5))
df['label'].value_counts().plot(kind='bar', color='teal')
plt.title('Crop Class Distribution - Check for Imbalance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../reports/crop_distribution.png')
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('../reports/correlation_heatmap.png')
```

```python
fig, axes = plt.subplots(2, 5, figsize=(20,8))
for i, col in enumerate(['N','P','K','ph','moisture','temperature','humidity','rainfall','ec','organic_carbon']):
    if col in df.columns:
        df[col].hist(ax=axes[i//5][i%5], bins=30, color='steelblue', edgecolor='white')
        axes[i//5][i%5].set_title(col)
plt.tight_layout()
plt.savefig('../reports/feature_distributions.png')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,6))
for ax, col in zip(axes, ['N','P','K']):
    df.boxplot(column=col, by='label', ax=ax)
    ax.set_title(f'{col} by Crop')
    ax.set_xlabel('')
plt.tight_layout()
plt.savefig('../reports/npk_by_crop.png')
```

```python
counts = df['label'].value_counts()
imbalance_ratio = counts.max() / counts.min()
print(f'Imbalance ratio: {imbalance_ratio:.2f}')
if imbalance_ratio > 2:
    print('APPLY SMOTE during training!')
```
